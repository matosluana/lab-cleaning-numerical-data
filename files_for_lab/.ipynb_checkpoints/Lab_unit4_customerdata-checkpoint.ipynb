All the columns with <10 uniques will be treated as discrete. All columns with thousands of values are continuous.
I need a closer look at the columns inbetween to make a decision. Before that, I will separate the categorical columns and differentiate only the numerical columns between continuous and discrete.

```python
cat_df = customer_df.select_dtypes('object')
customer_num = customer_df.drop(columns=cat_df.columns)

display(cat_df.head())
display(customer_num.head())
```

```python
discrete_columns = []
for col in customer_num:
    if customer_num[col].nunique() <10:
        discrete_columns.append(col)
    else:
        pass

discrete_columns
```

```python
continuous_columns = []
for col in customer_num:
    if customer_num[col].nunique() >1000:
        continuous_columns.append(col)
    else:
        pass

continuous_columns
```

```python
copy = customer_num.drop(columns=discrete_columns)
copy = copy.drop(columns=continuous_columns)
copy.describe(include='all').T
```

```python
# I will treat effective_to_date as discrete, since it can be only be days from January or March 2011
copy['effective_to_date'].unique() 
```

```python
copy['monthly_premium_auto'].unique()
# I will treat as continuous because too many different values.
# Even though these are all integers, I will assume those can be any number in the range, including non integers
```

```python
# Those are all entire numbers ranging from 0 to 35 months. I will treat them as discrete
copy['months_since_last_claim'].unique() 
```

```python
# Those are all entire numbers ranging from 0 to 99 months. I will treat them as discrete
copy['months_since_policy_inception'].unique()
```

```python
# After taking these decisions, any columns with > 200 values can be treated as continuous for this dataset

def discrete(dataframe, limit_uniques):
    discrete_columns = []
    for col in dataframe:
        if dataframe[col].nunique() <= limit_uniques:
            discrete_columns.append(col)
        else:
            pass
    discrete_df = dataframe[discrete_columns]
    return discrete_df

discrete_df = discrete(customer_num, 200)
discrete_df
```

```python
continuous_df = customer_num.drop(columns=discrete_df.columns)
continuous_df.head()
```

### Checking for outliers and NaNs

```python
# The boxplot suggests that there are outliers in the numerical data, except for income. I will check with a function

def find_outliers(dataframe):
    q1 =dataframe.quantile(0.25)
    q3=dataframe.quantile(0.75)
    IQR=q3-q1
    outliers = dataframe[((dataframe<(q1-1.5*IQR)) | (dataframe>(q3+1.5*IQR)))]

    return outliers


for col in continuous_df:
    outliers = find_outliers(continuous_df[col])
    print(col)
    print('number of outliers: '+ str(len(outliers)))
    print('max outlier value: '+ str(outliers.max()))
    print('min outlier value: '+ str(outliers.min()))
    outliers
```

## X-y and train-test split

```python
X = df.drop('total_claim_amount', axis=1)
y = df['total_claim_amount']

print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Encoding of categoricals

```python
categoricals_train= X_train.select_dtypes(object)
categoricals_test= X_test.select_dtypes(object)
categoricals_train.head()
```

```python
for column in categoricals_train.columns:
    print(column)
    print(categoricals_train[column].unique())
```

```python
to_one_hot = ['state', 'response', 'gender', 'marital_status', 'policy_type', 'policy',
       'renew_offer_type', 'sales_channel', 'vehicle_class']
to_ordinal = ['coverage', 'education', 'employment_status','location_code', 'vehicle_size']
```

```python
X_train_to_1hot = X_train[to_one_hot]
X_test_to_1hot = X_test[to_one_hot]
```

```python
encoder = OneHotEncoder(drop='first').fit(X_train_to_1hot)
cols = encoder.get_feature_names_out(input_features = to_one_hot)
X_train_1hot_encoded = pd.DataFrame(encoder.transform(X_train_to_1hot).toarray(), columns=cols).reset_index(drop=True)
X_test_1hot_encoded = pd.DataFrame(encoder.transform(X_test_to_1hot).toarray(), columns=cols).reset_index(drop=True)
print(X_train_1hot_encoded.shape)
print(X_test_1hot_encoded.shape)
```

```python
X_test_1hot_encoded
```

```python
X_train_to_ordinal = X_train[to_ordinal].reset_index(drop=True)
X_test_to_ordinal = X_test[to_ordinal].reset_index(drop=True)

X_train_ordinal_encoded = X_train_to_ordinal
X_test_ordinal_encoded = X_test_to_ordinal
X_train_ordinal_encoded
```

```python
X_train_ordinal_encoded["coverage"] = X_train_to_ordinal["coverage"].map({"Basic" : 0, "Extended" :.5, "Premium" : 1})
X_train_ordinal_encoded["education"] = X_train_to_ordinal["education"].map({"High School or Below" : 0, "Bachelor" : .5,"Postgraduate" : 1})
X_train_ordinal_encoded["employment_status"] = X_train_to_ordinal["employment_status"].map({"Medical Leave" : 0, "Disabled" : 0, "Unemployed" : 0, "Retired" : .5, "Employed":1})
X_train_ordinal_encoded["location_code"] =X_train_to_ordinal["location_code"].map({"Rural" : 0, "Suburban" :.5, "Urban" : 1,})
X_train_ordinal_encoded["vehicle_size"] =X_train_to_ordinal["vehicle_size"].map({"Small" : 0, "Medsize" :.5, "Large" : 1,})
X_train_ordinal_encoded
```

```python
X_test_ordinal_encoded["coverage"] = X_test_to_ordinal["coverage"].map({"Basic" : 0, "Extended" :.5, "Premium" : 1})
X_test_ordinal_encoded["education"] = X_test_to_ordinal["education"].map({"High School or Below" : 0, "Bachelor" : .5,"Postgraduate" : 1})
X_test_ordinal_encoded["employment_status"] = X_test_to_ordinal["employment_status"].map({"Medical Leave" : 0, "Disabled" : 0, "Unemployed" : 0, "Retired" : .5, "Employed":1})
X_test_ordinal_encoded["location_code"] =X_test_to_ordinal["location_code"].map({"Rural" : 0, "Suburban" :.5, "Urban" : 1,})
X_test_ordinal_encoded["vehicle_size"] =X_test_to_ordinal["vehicle_size"].map({"Small" : 0, "Medsize" :.5, "Large" : 1,})
X_test_ordinal_encoded
```

```python
X_train_encoded = pd.concat([X_train_1hot_encoded, X_train_ordinal_encoded], axis=1)
X_train_encoded
```

```python
X_test_encoded = pd.concat([X_test_1hot_encoded, X_test_ordinal_encoded], axis=1)
X_test_encoded
```

## Scaling the numericals

```python
numericals_train= X_train.select_dtypes(np.number).reset_index(drop=True)
numericals_test= X_test.select_dtypes(np.number).reset_index(drop=True)
```

```python
transformer = MinMaxScaler().fit(numericals_train)
cols_num = numericals_train.columns
numericals_train_scaled = pd.DataFrame(transformer.transform(numericals_train), columns=cols_num)
numericals_test_scaled = pd.DataFrame(transformer.transform(numericals_test), columns=cols_num)
```

## Concatenate numericals and categoricals

```python
X_train_transformed = pd.concat([numericals_train_scaled,X_train_encoded],axis=1)
X_test_transformed = pd.concat([numericals_test_scaled,X_test_encoded],axis=1)
```

```python
X_train_transformed
```

```python
X_test_transformed
```

```python
X_train_transformed.dtypes
```

```python
X_test_transformed.dtypes
```

## Compare models

```python
import warnings
warnings.filterwarnings('ignore')


model1 = LinearRegression()

model2 = KNeighborsRegressor(n_neighbors=3)

model3= SGDRegressor()


model_pipeline = [model1, model2, model3]
model_names = ['Linear Regression', 'KNN', 'SGDRegressor']
for model, model_name in zip(model_pipeline, model_names):
    model.fit(X_train_transformed, y_train)
    print(model_name)
    print('Train score:',model.score(X_train_transformed, y_train))
    print('Test score:',model.score(X_test_transformed, y_test),'\n')
```

The log transformation has indeed helped to approximate total claim amount and customer lifetime value to a normal distribution.
